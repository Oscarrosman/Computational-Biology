# Problem 2 - Discrete growth models



### Part a)

Find fixed points for non-negative steady states

K, r > 0, b >= 1

```python
from sympy import symbols, Eq, solve, diff
import numpy as np
import matplotlib.pyplot as plt
```

```python


N, t = symbols('N_tau tau', nonnegative=True)

r, K = symbols('r K', positive=True)  # K > 0, r > 0

b = symbols('b', positive=True, bounded=(1, None))  # b >= 1

equation = Eq(N, (r + 1) * N / (1 + (N / K)**b))

solutions = solve(equation, N)

for solution in solutions:
    print(f"N(t) = {solution}")

```

### Part b)

Perform linear stability analysis, $|f'| < 1$ for stability.

```python
equation = (r + 1) * N / (1 + (N / K)**b)
derivative = diff(equation, N)
linearized = [derivative.subs(N, solutions[0]), derivative.subs(N, solutions[1])]
for i, l in enumerate(linearized):
    print(f"({i+1}) Linearized: {l}")
```

I) Looking at the linear stability at point (1) where N = 0 it is clear that the point cannot be stable while fufilling the initial constraint r > 0

II) The constraints on the parameters at linearization (2) is harder to determine since they depend on both b and r. Further discussion on the balance between these parameters is done in (c)

### Part c)

Analytically find all conditions on the parameters within their valid range where a bifurcation from steady to unsteady state occurs.

Solution: A continuation of part (b). The bifurcation occurs at the fixed point where the solution goes from a steady state (|N'| < 1) to an unsteady state (|N'| > 1). The code below then solves for |N'| = 1 to find the condition where the bifurcation occurs.

```python
# Define and differentiate the equation
equation = (r + 1) * N / (1 + (N / K)**b)
NDot = equation.diff(N)

# Insert the fixed point into the derivative
fp = solutions[1]
NDotFP = NDot.subs(N, fp)
print(f'f\'(FP2) = {NDotFP}')

# Define and solve the equation for stability condition <= 1
eqn = Eq(abs(NDotFP), 1)
solution = solve(eqn,r)
print(f'r = {solution[0]}')
```

In this task only fixed point II is considered since fixed point I is always unstable. To find the transition from stable to unstable the function is linearized at the fixed point and set equal to 1, where the transition from stable to unstable occurs. At this point we solve the equation for r to set a dependency to b.

When solving for r we get the following expression: $r = \frac{2}{b - 2}$
This come with the additional constraint that it is undefined for $b = 2$. Since r is the growth rate and must be larger than zero $b < 2$ is not an accepted solution either. This then results in the constraint $b > 2$. 

### Part d)

K = 10^3, r = 0.1, b = 1

Investigate how the population size depends on time for the following values of the initial population size: N0 = 1, 2, 3 and 10. For each case, use the linear stability analysis performed in task b) to approximate the population dynamics in the vicinity of the unstable steady state.

```python
NInitial = [1,2,3,10]
K = 10**3
b = 1
r = 0.1
linearized = [r+1, -b*r/(r + 1) + 1]
solutions = [0, K*r**(1/b)]

Ntau = lambda N: (r + 1) * N / (1 + (N / K)**b)
lin1 = lambda N: linearized[0] * (N)

popApproximation = [[N0] for N0 in NInitial]
popLinearApproximation1 = [[N0] for N0 in NInitial]
t = 100

for startCondition in range(len(NInitial)):
    for _ in range(t):
        popApproximation[startCondition].append(Ntau(popApproximation[startCondition][-1]))
        popLinearApproximation1[startCondition].append(lin1(popLinearApproximation1[startCondition][-1]))

t = np.linspace(0, t, t+1)
colors = ['r', 'g', 'b', 'black']

# Plot linearization 1 (Unstable)
plt.title('Population dynamics around the unstable fixed point, N* = 0')
for i in range(len(NInitial)):
    plt.loglog(t, popApproximation[i], color=colors[i],label=f"N0 = {NInitial[i]}")
    plt.loglog(t, popLinearApproximation1[i], color=colors[i], linestyle='--',label=f"Linearized (N0 = {NInitial[i]})")
    plt.xlim(1, 100)
    plt.ylim(1, 1000)
plt.xlabel('Time')
plt.ylabel('Population')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.77))
```

### Part e)

The linear approximation follows the function well initially but diverges as the function reaches its stable fixed point. This is expected as the linearization is made around a single point and the function is not linear the quality of the approximation will deteriorate further from the point. 

From the figure it appears that for the smaller initial conditions the linearization is more stable for longer and diverges later.

### Part f)

```python
dN0 = [-10, -3, -2, -1, 1, 2, 3, 10]
NInitial = [solutions[1] + d for d in dN0]
colors = ['r', 'g', 'b', 'black', 'r', 'g', 'b', 'black']

lin2 = lambda N: solutions[1] + linearized[1] * (N - solutions[1])
#lin2 = lambda N: linearized[1] * (N)
popApproximation = [[N0] for N0 in NInitial]
popLinearApproximation2 = [[N0] for N0 in NInitial]
for startCondition in range(len(NInitial)):
    for _ in range(100):
        popApproximation[startCondition].append(Ntau(popApproximation[startCondition][-1]))
        popLinearApproximation2[startCondition].append(lin2(popLinearApproximation2[startCondition][-1]))

plt.title('Population dynamics around the stable fixed point, N* = -b*r/(r + 1) + 1')
for i in range(len(NInitial)):
    plt.loglog(t, popApproximation[i], color=colors[i], label=f"N0 = {NInitial[i]}")
    plt.loglog(t, popLinearApproximation2[i], color=colors[i], linestyle='--',label=f"Linearized (N0 = {NInitial[i]})")
    plt.xlim(1, 100)
    plt.ylim(85, 115)
    
plt.legend(loc='center left', bbox_to_anchor=(1, 0.53))
plt.xlabel('Time')
plt.ylabel('Population size')
plt.show()
```
