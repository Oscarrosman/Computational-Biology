```python
from tqdm import tqdm
alpha = 2
beta = 1
iterations = 10**4
tMax = 4000
dt = 0.1
n = np.zeros((iterations, int(tMax/dt)))
N = 30
n0 = N*(1-beta/alpha)
n[:,0] = n0

with tqdm(total=iterations) as pbar:
    for iteration in range(iterations):
        lastIdx = 0
        t = 0

        while t < tMax-1:
            if n[iteration, lastIdx] < 1:
                break
            
            bn = alpha*n[iteration, lastIdx]*(1-n[iteration, lastIdx]/N)
            dn = beta*n[iteration, lastIdx]
            tb = np.random.exponential(1/bn)
            td = np.random.exponential(1/dn)
            t += min([tb, td])
            idx = int(np.floor(t/dt)+1)
            n[iteration, lastIdx:idx] = n[iteration, lastIdx]
            if tb < td:
                n[iteration, idx] += 1
            else:
                n[iteration, idx] -= 1
            lastIdx = idx

        
        pbar.update(1)

Pn = np.zeros_like(n[0,:])
for i in range(len(Pn)):
    Pn[i] = np.sum(n[:,i])/iterations


task = 0
fontsize = 15
bins = 50
figsize = 6
T_list = np.array([50, 100, 150])

#np.save('save/n.npy', n)
# n = np.load('n.npy')

t_list = np.arange(0, tMax, dt)

fig, ax1 = plt.subplots(figsize=(figsize,figsize))
ax1.plot(t_list, Pn)
# ax1.set_title('$T_\\{max} = {}$'.format(T_list[task]), fontsize=fontsize)
# ax1.set_yscale('log')
# ax1.set_ylabel('$logP(n_t)$', fontsize=fontsize)
# ax1.set_xlabel('$t$', fontsize=fontsize)
ax1.set_box_aspect(1) 
ax1.legend(loc="upper left", prop={'size': 11})

# plt.legend(loc="upper right", prop={'size': 11})
title = '/3.1d_T_{}'.format(T_list[task])
# plt.savefig(location+title+'.png')
plt.show()
```
